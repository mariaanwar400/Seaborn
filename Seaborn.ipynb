```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate some sample data (like restaurant bills or exam scores)
data = np.random.normal(loc=50, scale=10, size=200)

# -----------------------------------------------------
# 1. Histogram (basic view of data distribution)
# -----------------------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(data, bins=20, color="skyblue")
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()

# -----------------------------------------------------
# 2. Histogram + KDE (smooth distribution curve)
# -----------------------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(data, bins=20, kde=True, color="lightgreen")
plt.title("Histogram + KDE Curve")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# -----------------------------------------------------
# 3. KDE Only (smooth curve without histogram)
# -----------------------------------------------------
plt.figure(figsize=(8,5))
sns.kdeplot(data, shade=True, color="red")
plt.title("KDE Curve Only")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# -----------------------------------------------------
# 4. Histogram + Normal Distribution Fit
# -----------------------------------------------------
plt.figure(figsize=(8,5))
sns.distplot(data, bins=20, fit=norm, color="purple")  # distplot is deprecated but useful for demo
plt.title("Histogram + Fitted Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# -----------------------------------------------------
# 5. Comparison: KDE vs Normal Fit
# -----------------------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(data, bins=20, stat="density", color="lightgray")

# KDE
sns.kdeplot(data, color="blue", label="KDE (from data)")

# Normal distribution curve
mu, sigma = np.mean(data), np.std(data)
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mu, sigma), color="red", linestyle="--", label="Normal Fit")

plt.title("KDE vs Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    "age": [22, 25, 30, 35, 40],
    "income": [20000, 30000, 40000, 45000, 60000],
    "score": [50, 60, 65, 70, 80]
}
df = pd.DataFrame(data)

# Correlation matrix
print(df.corr())

# Heatmap visualization
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

How to read:

Values range from -1 to 1.

+1 â†’ perfect positive correlation (as one increases, the other also increases).

-1 â†’ perfect negative correlation (as one increases, the other decreases).

0 â†’ no linear correlation.
Example: 0.85 means strong positive relation.

```python
import pandas as pd
import scipy.stats as stats
import numpy as np

# Example DataFrame
df = pd.DataFrame({
    "gender": ["M", "F", "F", "M", "M"],
    "purchased": ["Yes", "No", "Yes", "Yes", "No"]
})

# Crosstab
table = pd.crosstab(df["gender"], df["purchased"])
chi2 = stats.chi2_contingency(table)[0]
n = table.sum().sum()
min_dim = min(table.shape)-1

# Cramer's V
cramers_v = np.sqrt(chi2 / (n * min_dim))
print("Cramer's V:", cramers_v)
```

How to read:

0 = no association.

0.1â€“0.3 = weak, 0.3â€“0.5 = moderate, >0.5 = strong association.

```python
from scipy.stats import pointbiserialr

# Example
df = pd.DataFrame({
    "gender": [0, 1, 1, 0, 0],  # 0 = Male, 1 = Female
    "income": [20000, 30000, 40000, 25000, 27000]
})

corr, pval = pointbiserialr(df["gender"], df["income"])
print("Correlation:", corr, "p-value:", pval.round(3))
```
